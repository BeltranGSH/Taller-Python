## 1. Entrada y Salida de Datos en Python

En esta sección trabajamos con la función `input()` para capturar datos del usuario y la conversión de tipos de datos.

## 3. Estructuras de Control

Uso de condicionales (if/elif/else) para tomar decisiones en el programa.

## 4. Bucles y Estructuras de Repetición

Implementación de bucles for y while para repetir operaciones.

```python
# Bucle FOR: Tabla de multiplicar
numero = int(input("¿De qué número quieres la tabla de multiplicar? "))

print(f"\nTabla del {numero}:")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

```python
# Bucle WHILE: Suma de números hasta llegar a un límite
suma = 0
contador = 1
limite = int(input("Ingrese el límite: "))

while contador <= limite:
    suma += contador
    contador += 1

print(f"La suma de los números del 1 al {limite} es: {suma}")
```

## 5. Listas y Estructuras de Datos

Trabajo con listas para almacenar y manipular colecciones de datos.

```python
# Crear y manipular una lista de números
numeros = [12, 45, 23, 67, 34, 89, 15, 56]

print("Lista original:", numeros)
print(f"Primer elemento: {numeros[0]}")
print(f"Último elemento: {numeros[-1]}")
print(f"Cantidad de elementos: {len(numeros)}")
print(f"Número mayor: {max(numeros)}")
print(f"Número menor: {min(numeros)}")
print(f"Suma total: {sum(numeros)}")
print(f"Promedio: {sum(numeros)/len(numeros):.2f}")
```

```python
# Filtrar números pares e impares
pares = []
impares = []

for num in numeros:
    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)

print("Números pares:", pares)
print("Números impares:", impares)
```

## 6. Funciones

Creación de funciones reutilizables para organizar mejor el código.

```python
# Función para calcular el área de un círculo
def calcular_area_circulo(radio):
    """Calcula el área de un círculo dado su radio"""
    return math.pi * radio ** 2

# Función para calcular el volumen de una esfera
def calcular_volumen_esfera(radio):
    """Calcula el volumen de una esfera dado su radio"""
    return (4/3) * math.pi * radio ** 3

# Probar las funciones
r = 5
print(f"Radio = {r}")
print(f"Área del círculo: {calcular_area_circulo(r):.2f}")
print(f"Volumen de la esfera: {calcular_volumen_esfera(r):.2f}")
```

## 11. Aplicación Práctica: Movimiento Rectilíneo Uniforme (MRU)

Ejemplo práctico de física aplicando conceptos de programación.

```python
# Movimiento Rectilíneo Uniforme: d = v * t
def calcular_mru(velocidad, tiempo):
    """Calcula la distancia en un MRU"""
    return velocidad * tiempo

# Datos del problema
v = 20  # m/s
tiempo = np.linspace(0, 10, 100)  # de 0 a 10 segundos
distancia = calcular_mru(v, tiempo)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(tiempo, distancia, linewidth=2, color='green')
plt.xlabel('Tiempo (s)')
plt.ylabel('Distancia (m)')
plt.title(f'Movimiento Rectilíneo Uniforme (v = {v} m/s)')
plt.grid(True, alpha=0.3)
plt.show()

print(f"A los 5 segundos, la distancia recorrida es: {calcular_mru(v, 5)} metros")
```

## 12. Aplicación Práctica: Movimiento Rectilíneo Uniformemente Acelerado (MRUA)

Cálculo de posición y velocidad con aceleración constante.

```python
# MRUA: x = x0 + v0*t + (1/2)*a*t^2
def calcular_mrua(x0, v0, a, t):
    """Calcula la posición en un MRUA"""
    return x0 + v0*t + 0.5*a*t**2

def calcular_velocidad_mrua(v0, a, t):
    """Calcula la velocidad en un MRUA"""
    return v0 + a*t

# Datos
x0 = 0  # posición inicial (m)
v0 = 10  # velocidad inicial (m/s)
a = 2  # aceleración (m/s^2)
t = np.linspace(0, 10, 100)

posicion = calcular_mrua(x0, v0, a, t)
velocidad = calcular_velocidad_mrua(v0, a, t)

# Graficar ambas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(t, posicion, color='blue', linewidth=2)
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Posición (m)')
ax1.set_title('Posición vs Tiempo')
ax1.grid(True, alpha=0.3)

ax2.plot(t, velocidad, color='red', linewidth=2)
ax2.set_xlabel('Tiempo (s)')
ax2.set_ylabel('Velocidad (m/s)')
ax2.set_title('Velocidad vs Tiempo')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
